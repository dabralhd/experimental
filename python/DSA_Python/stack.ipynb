```python
from builtins import Exception

class Stack:
    def __init__(self):
        self._data = list()

    def push(self, elem):
        self._data.append(elem)

    def is_empty(self):
        return len(self._data) == 0
    
    def pop(self):
        if self.is_empty():
            raise Exception('underflow')
        top = self._data[-1]
        self._data.pop()

    def top(self):
        if self.is_empty():
            raise Exception('underflow')
        return self._data[-1]
    


```

```python
s = Stack()
for x in range(1,11,2):
    s.push(x)

```

```python
try:
    print(f'top: {s.top()}')
    s.pop()
except Exception as e:
    print(f'exception occurred: {e}')
else:
    print('stack pop - ok')
    

```
